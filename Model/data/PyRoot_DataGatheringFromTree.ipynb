```python
import csv
import re
import os
eventBranches = list(eventTree.keys())
os.mkdir("dataFolder")
# writer.writerow(eventBranches)
for i in eventBranches:
  temp = eventTree[i].array()
  tempCsv = None
  nameCsv =  re.sub('\W+','', i)

  tempCsv = open("dataFolder/"+nameCsv+".csv", 'w', newline='')  

  writer = csv.writer(tempCsv)
  writer.writerow(temp)
  tempCsv.close()
```
